# Data Preprocessing

```python
loan_data = loan_data.drop("loan_id" , axis = 1)
```

```python
loan_data.select_dtypes(['object'])
```

## Encode categorical columns

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
for col in loan_data.select_dtypes(include = "object").columns:
    loan_data[col] = encode.fit_transform(loan_data[col])
```

```python
loan_data.head()
```

## Check class Imbalance

```python
loan_data.columns
```

```python
loan_data[" loan_status"].value_counts()
```

# Data Split into Train and Test

```python
x = loan_data.drop([" loan_status"] , axis = 1 )
x
```

```python
y = loan_data[" loan_status"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train,y_test = train_test_split(x , y , test_size = 0.3 , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion = "entropy", max_depth = 5 , min_samples_split = 10 , random_state = 42, )
tree_model.fit(x_train , y_train)
```

```python
y_pred = tree_model.predict(x_test)
```

# Evaluate Model

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

# SMOTE to Address class Imbalance 

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state = 42)
x_train_smote , y_train_smote = smote.fit_resample(x_train,y_train)

print(f"Before Smote: {y_train.value_counts()}")
print(f"After Smote: {y_train_smote.value_counts()}")
```

# Train Logistic Regression On Balance Data

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(max_iter = 1000 , random_state = 42)
lr_model.fit(x_train_smote , y_train_smote)
```

```python
y_pred_smote = lr_model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_smote))
```

# Bonus Experiment – SMOTE + Logistic Regression vs. Decision Tree
**A Decision Tree classifier trained on the original imbalanced dataset achieved very high accuracy (97%) with strong precision and recall for both classes, indicating it handled the imbalance well.
When SMOTE was applied to balance the training data and Logistic Regression was used, performance on the original test set dropped to 79% accuracy, with noticeably lower precision and recall. This suggests that in this dataset, synthetic oversampling with a linear model introduced noise and reduced generalization, whereas the non-linear Decision Tree was more capable of capturing the patterns without balancing.**
